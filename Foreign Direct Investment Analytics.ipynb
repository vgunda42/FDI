```python
# Plotting FDI trends over the years for each sector
plt.figure(figsize=(15, 8), facecolor='white')
for sector in df['Sector'].unique():
    plt.plot(df.columns[1:], df[df['Sector'] == sector].iloc[0, 1:], marker='o', label=sector)
plt.title('FDI Trends by Sector Over Years')
plt.xlabel('Year')
plt.ylabel('FDI (in millions)')
plt.legend(title='SECTOR', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df['Total FDI'] = df.iloc[:, 1:].sum(axis=1)

# Sorting the sectors by total FDI
top_sectors = df[['Sector', 'Total FDI']].sort_values(by='Total FDI', ascending=False)

# Displaying the top sectors
print(top_sectors.head(10))
```

```python
# Extracting FDI data for the Automobile Industry
df_automobile = df[df['Sector'] == 'AUTOMOBILE INDUSTRY'].iloc[:, 1:-1]  # Exclude the 'Total FDI' column

# Plotting FDI trend for the Automobile Industry
plt.figure(figsize=(12, 6), facecolor='white')
plt.plot(df.columns[1:-1], df_automobile.iloc[0], marker='o', linestyle='-', color='b')
plt.title('FDI Trend in Automobile Industry Over Years')
plt.xlabel('Year')
plt.ylabel('FDI (in millions)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculating the correlation matrix for different sectors based on their FDI values over the years
# First, we need to create a new DataFrame where each sector's FDI values are columns
sectors_fdi = df.drop(columns=['Total FDI']).set_index('Sector').T
sectors_correlation = sectors_fdi.corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 10), facecolor='white')
sns.heatmap(sectors_correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Between Different Sectors')
plt.xlabel('SECTOR')
plt.ylabel('SECTOR')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

```python
# Calculating the proportion of total FDI that each sector contributes
total_fdi = df['Total FDI'].sum()
df['FDI Proportion'] = df['Total FDI'] / total_fdi

# Plotting the proportions as a pie chart
plt.figure(figsize=(10, 8), facecolor='white')
plt.pie(df['FDI Proportion'], labels=df['Sector'], autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Total FDI by Sector')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Summing up the FDI for each sector over the past decade to find the top sectors
df['Total FDI Last Decade'] = df[df.columns[-11:-1]].sum(axis=1)
top_sectors = df[['Sector', 'Total FDI Last Decade']].sort_values(by='Total FDI Last Decade', ascending=False).head(3)
print(top_sectors)
```
